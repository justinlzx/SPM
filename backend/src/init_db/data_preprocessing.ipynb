```python
import pandas as pd
import random

# Load the CSV file
file_path = "latest_arrangement.csv"  # Update this path if necessary
df = pd.read_csv(file_path)

# Define the reasons and prefixes
reasons = [
    "Personal health issues, dental or specialist appointments, routine check-ups, or recovery after minor surgery.",
    "Taking care of a family member, elderly care, or childcare issues.",
    "Taking a day for self-care or to manage stress and prevent burnout.",
    "Handling important personal matters such as legal appointments, bank work, or home repairs.",
    "Attending a family wedding, birthday, or significant milestone celebration.",
    "Planned time off to recharge, travel, or enjoy a break.",
    "Taking leave to attend a course, certification, or exam preparation.",
    "Moving to a new home or handling logistics related to a recent move.",
    "Celebrating a religious holiday or participating in a faith-based event.",
    "Weather disruptions, transportation issues, or emergencies.",
]


# Function to randomly assign WFH or OOO with 75% for WFH
def random_reason():
    prefix = "WFH - " if random.random() < 0.75 else "OOO - "
    reason = random.choice(reasons)
    return prefix + reason


# Populate column I with random reasons
df["I"] = df.apply(lambda row: random_reason(), axis=1)

# Save the updated CSV
df.to_csv("latest_arrangement_reason.csv", index=False)
print("Column I populated and saved to 'latest_arrangements_populated.csv'")
```
